```python
a = []
b = []
c = []
for i in T.index:
    a.extend([X[X['CUSTOMER_ID'] == i].to_numpy()[:,1]])
    b.extend([X[X['CUSTOMER_ID'] == i].to_numpy()[:,2]])
    c.extend([X[X['CUSTOMER_ID'] == i].to_numpy()[:,3]])
```

```python
e = []
f = []
g = []
for i in range(len(a)):
    if len(a[i])>40 and len(a[i])<50:
        e.append(a[i])
        f.append(b[i])
        g.append(c[i])
    elif len(a[i])>= 50:
        e.append(a[i][0:49])
        f.append(b[i][0:49])
        g.append(c[i][0:49])
```

```python
padded_inputs1 = tf.keras.preprocessing.sequence.pad_sequences(
    e, padding="post", maxlen=None, dtype='int32', value=-1.0)
padded_inputs2 = tf.keras.preprocessing.sequence.pad_sequences(
    f, padding="post", maxlen=None, dtype='float32', value=-1.0)
padded_outputs = tf.keras.preprocessing.sequence.pad_sequences(
    g, padding="post", maxlen=None, dtype='int32', value=1.0)
print(padded_inputs1.shape)
print(padded_inputs2.shape)
print(padded_outputs.shape)
print(5658*49)
```

```python
x_train1 = padded_inputs1.reshape(-1,1,49)
x_train2  = padded_inputs2.reshape(-1,1,49)
y_train  = padded_outputs.reshape(-1,1,49)
print(y_train.shape)
```

```python
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Masking,Dense,Bidirectional,Input,LSTM,Embedding,SimpleRNN
```

```python
monto_input = tf.keras.Input(shape=(1,49,), name="monto")

receptor_input = tf.keras.Input(shape=(1,49,), name="receptor")

monto_capa = tf.keras.layers.Masking(mask_value=-1.0)(monto_input)

receptor_capa = tf.keras.layers.Masking(mask_value=-1.0)(receptor_input)

monto_capa = tf.keras.layers.Bidirectional(tf.keras.layers.LSTM(49,return_sequences =True, input_shape = (1,49,)))(monto_capa)

receptor_capa = tf.keras.layers.Bidirectional(tf.keras.layers.LSTM(49,return_sequences =True, input_shape = (1,49,)))(receptor_capa)

x = tf.keras.layers.average([monto_capa, receptor_capa])

#x = tf.keras.layers.Dense(49)(x)

#x= tf.keras.layers.LSTM(49,return_sequences=True)(x)

x = tf.keras.layers.TimeDistributed(tf.keras.layers.Dense(49))(x)

clase_pred = tf.keras.layers.Dense(49, activation='sigmoid' ,name="clase")(x)

model = tf.keras.Model(
    inputs=[monto_input,receptor_input],
    outputs=[clase_pred],
)
```

```python
model.compile(
    optimizer=tf.keras.optimizers.Adam(),
    loss=[tf.keras.losses.BinaryCrossentropy()],
    metrics=['accuracy'],
)
```

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import sklearn
from sklearn.metrics import confusion_matrix
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import seaborn as sns
```

```python
def plot_cm(labels, predictions, p=9.9997638e-01):
  cm = confusion_matrix(labels, predictions > p)
  plt.figure(figsize=(5,5))
  sns.heatmap(cm, annot=True, fmt="d")
  plt.title('Matriz de confusion @{:.2f}'.format(p))
  plt.ylabel('Valor actual')
  plt.xlabel('Valor predecido')

  print('DetecciÃ³n de una transacciÃ³n legÃ­tima (Verdaderos Positivos): ', cm[0][0])
  print('TransacciÃ³n legÃ­tima incorrectamente detectatada (Falso Negativo): ', cm[0][1])
  print('TransacciÃ³n fraudulenta incorrectamente detectada (Falso Positivo): ', cm[1][0])
  print('TransacciÃ³n fraudulenta detectada: ', cm[1][1])
  print('Total de transacciones del titular: ', np.sum(cm[1]))
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import make_classification
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import roc_curve

def plot_roc_curve(fper, tper):
    plt.plot(fper, tper, color='red', label='ROC')
    plt.plot([0, 1], [0, 1], color='green', linestyle='--')
    plt.xlabel('Falsos Positivos')
    plt.ylabel('Verdaderos Positivos')
    plt.title('Transacciones Fraudulentas titular 5000')
    plt.legend()
    plt.show()
    
data_X, cls_lab = make_classification(n_samples=2100, n_classes=2, weights=[1.939,1.2959], random_state=2)
train_X, test_X, train_y, test_y = train_test_split(data_X, cls_lab, test_size=0.4, random_state=3)

model =RandomForestClassifier()
model.fit(train_X, train_y)

prob = model.predict_proba(test_X)
prob = prob[:, 1]
fper, tper, thresholds = roc_curve(test_y, prob)
plot_roc_curve(fper, tper)
```
